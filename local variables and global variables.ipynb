```python
n1=20
n2=30
def addition():           #no need to provide inside brackets
    #no need to mention
    add=n1+n2
    print('the addition of {} and {} is {}'.format(n1,n2,add))
    
addition()
```

```python
def avg():                                   #local variables
    num1=20
    num2=30
    num3=40
    avg=(num1+num2+num3)/3
    print('the average of numbers is:',avg)
    
avg()
```

```python
num11=20
num22=30
num33=40
def avg():                                   #global variables
    avg=(num1+num2+num3)/3
    print('the average of numbers is:',avg)
    return(num11)
    
val1=avg()
```

** local variables can be shown as global variables by using global keyword**

```python
n_=10
def updated():
    global n_
    n_=20
    n_=n_+100
    print('n1 value inside function is:',n_)
updated()
print(n_)

```
